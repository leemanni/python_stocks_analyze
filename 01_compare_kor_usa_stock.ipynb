# CJ 제일제당 vs MSFT

```python
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()

# CJ
cj = pdr.get_data_yahoo('097950.ks', start='2019-01-02')
cj_dpc = ((cj['Close'] - cj['Close'].shift(1))) / cj['Close'].shift(1) * 100
cj_dpc_cp = ((100 + cj_dpc) / 100).cumprod() * 100 - 100

# msft
msft = pdr.get_data_yahoo('msft', start='2019-01-02')
msft_dpc = ((msft['Close'] - msft['Close'].shift(1))) / msft['Close'].shift(1) * 100
msft_dpc_cp = ((100 + msft_dpc) / 100).cumprod() * 100 - 100

## make graph
import matplotlib.pyplot as plt

plt.plot(cj.index, cj_dpc_cp, 'b--', label='CJ')
plt.plot(msft.index, msft_dpc_cp, 'r', label='MSFT')
plt.ylabel('Change (%)')
plt.xlabel('date')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```
